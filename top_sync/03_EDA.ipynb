To visualize the infomation I have gathered, I'll be using Tableau. But I'll use this notebook to pull song quick stats from the dataset.

```python
import pandas as pd
import numpy as np
```

```python
songs = pd.read_csv('./data/all_songs_cleaned.csv')
```

```python
songs.dtypes
```

Average songs per episode of TV:

```python
songs[songs['type'] == 'TV']['avg_per_ep'].mean()
```

The average for "Clone High" wasn't calculated correctly so fixing that:

```python
songs.loc[songs['show'] == 'Clone High', 'avg_per_ep'] = 3.23
```

```python
songs[songs['type'] == 'TV'].sort_values('avg_per_ep', ascending = False).head(30)
```

```python
songs[songs['type'] == 'Movie']['avg_per_ep'].mean()
```

```python
songs[(songs['type'] == 'Movie') & (songs['show'] != 'How to Make It in America') & (songs['show'] != 'La La Land')].sort_values('avg_per_ep', ascending = False).head(30)
```

```python
songs.groupby('show')['avg_per_ep'].mean().sort_values(ascending=False)
```
