```python
y = [x for x in range(1, 4)] * 35
```

```python
x = []
for _ in range(1, 36):
    x.append(_)
    x.append(_)
    x.append(_)
```

```python
coord = [(x, y) for (x, y) in zip(x, y)]
```

```python
coord.pop(0)
```

```python
coord.pop(-3)
```

```python
coord = list(reversed(coord))
```

```python
def unpack_x(coordinate):
    return coordinate[0]
```

```python
def unpack_y(coordinate):
    return coordinate[1]
```

```python
races['coord'] = coord
```

```python
races['x'] = races['coord'].apply(unpack_x)
```

```python
races['y'] = races['coord'].apply(unpack_y)
```

```python
color = [x for x in range(0, 2)] * 52
```

```python
winners = drivers[drivers['finish'] <= 5]
```

```python
winners['height'] = 6 - winners['finish'] 
```

```python
winners.head()
```

```python
order_dict = {5: 1, 3:2, 1:3, 2:4, 4:5}
```

```python
winners['order'] = winners['finish'].map(order_dict)
```

```python
winners
```

```python
finish_dict = {1:'1st', 2:'2nd', 3:'3rd', 4:'4th', 5:'5th'}
```

```python
winners['place'] = winners['finish'].map(finish_dict)
```

```python
winners
```

```python
winners.to_csv('./tab_winners.csv', index = False)
```
